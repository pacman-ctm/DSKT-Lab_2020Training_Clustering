```python
import numpy as np 
import pandas as pd
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
# gender_submission = pd.read_csv("../input/titanic/gender_submission.csv")
```

```python
train_data['Age'].fillna(value = train_data['Age'].mean(), inplace = True)
train_data['Embarked'].fillna(value = 0, inplace = True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
colTrans = ColumnTransformer([
        ('scaler', StandardScaler(), ['Age','Fare'])
    ], remainder = 'passthrough')
data = pd.DataFrame(train_data)
data
```

```python
survived = train_data['Survived']
train_data.drop(['Cabin', 'Name', 'Ticket', 'PassengerId', 'Survived', 'Parch', 'SibSp'], axis = 1, inplace = True)
train_data.head()
train_data.dtypes
```

```python
data = colTrans.fit_transform(train_data)

data = pd.DataFrame(data)
data.head()
```

```python
data.columns = ['Age','Fare','Pclass','Sex','Embarked']
data.head()
```

```python
data.dtypes
```

```python
# Use map to change string to int.

data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'C': 0, 'S': 1, 'Q': 2})
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data["Fare"] = data["Fare"].astype(float)
data["Pclass"] = data["Pclass"].astype(int)
data['Embarked'].fillna(value = 0, inplace = True)
data['Age'].fillna(value = data['Age'].mean(), inplace = True)
data['Pclass'].unique()
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(data, survived)
train_x
```

```python
train_y
```

```python
data.isnull().sum()
```

# Support Vector Machine =))

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(train_x,train_y)
svm_model.score(test_x,test_y)
```
